Download and Visualize NYISO Actual Load and Forecast Data
============================================
This Jupyter Notebook shows an example on how to use the functionalities of this repository to download, organize and visualize NYISO data. Each magic function call `%run -i <script_name>.py` may be replaced by `from <script_name> import*` in a plain Python environment such as Spyder.

The first cell downloads and organizes the data and plots the actual load, forecast and forecast error for the previous day. 

```python
%run -i download_nyiso_data

# Downloading raw load forecast and actual load data from NYISO website
download_nyiso_data(2019, "00_NYISO_Data", print_info = True)
```

In the cell below, we plot the most recent load information. This cell may require a previous execution of the `download_nyiso_data` function to update the NYISO files to the most recent load/forecast version in the data server.

```python
%run -i visualize_load_forecast

import datetime

now = datetime.datetime.now() + datetime.timedelta(days = -1)

# Visualizing the actual load, load forecast
date = f"{now.month:02d}/{now.day:02d}/2020"

# NYC
time_stamp, load, forecast, worst_forecast = visualize_load_forecast(date, "N.Y.C.", 
                                                                     data_path = "00_NYISO_Data",
                                                                     show_plot = True)
# NYISO
time_stamp, load, forecast, worst_forecast = visualize_load_forecast(date, "NYISO",
                                                     data_path = "00_NYISO_Data",
                                                     show_plot = True)
```

```python
%run -i visualize_load_forecast

# NYISO
time_stamp, load, forecast, worst_forecast = visualize_load_forecast("01/03/2020", "NYISO",
                                                     data_path = "00_NYISO_Data",
                                                     show_plot = True)
```
