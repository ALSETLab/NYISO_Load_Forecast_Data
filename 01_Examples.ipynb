Download and Visualize NYISO Actual Load and Forecast Data
============================================
This Jupyter Notebook shows an example on how to use the functionalities of this repository to download, organize and visualize NYISO data. 

The first cell downloads and organizes the data and plots the actual load, forecast and forecast error for the previous day. 

```python
import datetime
from download_nyiso_data import*
from visualize_load_forecast import*

now = datetime.datetime.now()

# Downloading raw load forecast and actual load data from NYISO website
download_nyiso_data(2018, "00_NYISO_Data", print_info = True)
# Visualizing the actual load, load forecast
date = f"{now.month:02d}/{now.day-3:02d}/2020"
# NYC
time_stamp, load, forecast = visualize_load_forecast(date, "N.Y.C.", data_path = "00_NYISO_Data", show_plot = True)
# NYISO
time_stamp, load, forecast = visualize_load_forecast(date, "NYISO", data_path = "00_NYISO_Data", show_plot = True)
```
